# Task
Calculate the average age of all customers in the df DataFrame and then filter df to display only those customers whose age is less than this calculated average age.

## Calculate Average Age
Subtask:
Calculate the average age of all customers in the DataFrame using PySpark's avg aggregation function.

Reasoning: To calculate the average age, I need to import the avg function from pyspark.sql.functions, apply it to the 'age' column of the DataFrame, and then extract the numerical result.

```python
from pyspark.sql.functions import avg

avg_age_df = df.select(avg("age"))
avg_age = avg_age_df.collect()[0][0]

print(f"The average age of customers is: {avg_age}")
```

```python
df.filter(df.age < avg_age).show()
```

Bonus Exercise 1

Extract the first letter of each name and store in a new column

```python
from pyspark.sql.functions import col, substring

df_first_letter = df.withColumn("first_letter_name", substring(col("name"), 1, 1))
df_first_letter.show()
```

Bonus Exercise 2

Mask the customer_id. Example: C001 → ***01

```python
from pyspark.sql.functions import concat, lit, substring

df_masked_id = df.withColumn(
    "masked_customer_id",
    concat(lit("***"), substring(col("customer_id"), 4, 2))
)
df_masked_id.show()
```

Bonus Exercise 3

Create a new boolean column "is_senior" where age > 35

```python
from pyspark.sql.functions import when, col

df_with_is_senior = df.withColumn(
    "is_senior",
    when(col("age") > 35, True).otherwise(False)
)
df_with_is_senior.show()
```
