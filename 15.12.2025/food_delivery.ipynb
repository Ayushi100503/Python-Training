#SECTION B — JSON

Exercise 5

Write only “Delivered” orders to JSON.

```python
orders_df = spark.read \
.option("header", True) \
.option("inferSchema", True) \
.csv("/content/orders.csv")
```

```python
delivered_orders_df = orders_df.filter("order_status = 'Delivered'")
delivered_orders_df.write \
.mode("overwrite") \
.json("/content/delivered_orders.json")
delivered_orders_df.show()
```

Exercise 6

Read JSON and filter:
city = "Mumbai"
payment_mode = "Card"

```python
orders_json_df = spark.read.json("/content/delivered_orders.json")
filtered_df = orders_json_df.filter((orders_json_df.city == "Mumbai") &
                                    (orders_json_df.payment_mode == "Card"))
filtered_df.show()
```

Exercise 7

Add a column:
delivery_category
Logic:
delivery_time_minutes > 45 → "Late"
else → "OnTime"
Write output to JSON.

```python
from pyspark.sql.functions import when

final_df = filtered_df.withColumn("delivery_category",
                                        when(filtered_df.delivery_time_minutes > 45, "Late")
                                        .otherwise("OnTime"))
final_df.show()
```

Exercise 8

Force JSON output to a single partition and observe number of files created.

```python
final_df.coalesce(1).write \
.mode("overwrite") \
.json("orders_single_partition_json")
```

#SECTION C — PARQUET

Exercise 9

Convert full dataset to Parquet.
Output:
orders_parquet/

```python
orders_df = spark.read \
.option("header", True) \
.option("inferSchema", True) \
.csv("/content/orders.csv")
orders_df.write \
.mode("overwrite") \
.parquet("/content/orders.parquet")
```

Exercise 10

Read Parquet and filter:
cuisine = "Indian"
order_amount > 500

```python
orders_parquet_df = spark.read.parquet("/content/orders.parquet")
filtered_parquet_df = orders_parquet_df.filter((orders_parquet_df.cuisine == "Indian") &
                                                (orders_parquet_df.order_amount > 500))
filtered_parquet_df.show()
```

Exercise 11

Sort Parquet data by order_amount descending and write top 10 orders back to Parquet.

```python
top_10_orders_df = orders_parquet_df.orderBy(orders_parquet_df.order_amount.desc()).limit(10)
top_10_orders_df.write \
.mode("overwrite") \
.parquet("/content/top_10_orders.parquet")
```

#THINKING / ANALYTICS QUESTIONS

Exercise 15

Which cuisine generates the highest order_amount overall?

```python
from pyspark.sql.functions import sum
orders_df.groupBy("cuisine") \
.agg(sum("order_amount").alias("total_order_amount")) \
.orderBy("total_order_amount", ascending = False) \
.show(1)
```

Exercise 16

Which city has the highest number of orders?

```python
from pyspark.sql.functions import count
orders_df.groupBy("city") \
.agg(count("*").alias("order_count")) \
.orderBy("order_count", ascending = False) \
.show(1)
```

Exercise 17

Which payment mode is most frequently used?

```python
orders_df.groupBy("payment_mode") \
.count() \
.orderBy("count", ascending = False) \
.show(1)
```
